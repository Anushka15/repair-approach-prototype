```python
engine = create_engine_for_db()
```

```python
fd_constraint1 = [(['SID'],['SName'])]
fd_constraint1 = simplify_FDs(fd_constraint1)
```

```python
# Store all datasets in a single list


datasets = [df1,df2,df3,df4,df5,df6,df7,df8,df9,df10]

query_file_name_expected_CQA = "../queries/prototype1_truth.sql"
query_file_name_actual_CQA = "../queries/prototype1_CQA.sql"
query_file_name_expected_Prob = "../queries/prototype1_truth.sql"
query_file_name_actual_Prob = "../queries/prototype1_prob.sql"

total_precision_CQA, total_recall_CQA,total_coverage_CQA, total_noise_CQA = 0.0, 0.0, 0.0, 0.0
total_precision_prob, total_recall_prob,total_coverage_prob, total_noise_prob = 0.0, 0.0, 0.0, 0.0
sum_precision_p = [0.0] * 100
sum_recall_p = [0.0] * 100
sum_f1_p = [0.0] * 100
#sum_precision_p, sum_recall_p, sum_f1_p = 0.0, 0.0, 0.0
for i, df_data in enumerate(datasets, start=1):
    print(f"\n=== Dataset {i} ===")
    with engine.begin() as connection:
        connection.execute(text("""
        TRUNCATE TABLE prototype1.students_prob, prototype1.students
        RESTART IDENTITY
        CASCADE;
        """))

        # Reset/clear the DuBio dictionary for students
        connection.execute(text("""
        UPDATE prototype1."_dict"
        SET dict = dictionary('')
        WHERE name = :name;
        """), {"name": "students_dict"})
    df_data.to_sql('students',engine,schema='prototype1',index=False,if_exists='append')
    df_data = preprocess_data(df_data)
    rv_definitions,rv_probabilities,consistent_rows = create_rv_definitions(df_data,fd_constraint1)
    repaired_rows = create_repaired_rows(rv_definitions,df_data,consistent_rows)
    columns = [col for col in df_data.columns if col != 'uuid'] + ['_sentences']
    repaired_df = pd.DataFrame(repaired_rows, columns=columns)
    repaired_df = group_similar_rows_together(repaired_df)
    repaired_df.to_sql('students_prob',engine,schema='prototype1',index=False,if_exists='append')
    entries = [f"{k}:{v}" for k, v in rv_probabilities.items()]
    joined = ';'.join(entries)
    update_stmt = f"UPDATE prototype1._dict\nSET dict=add(dict, '{joined}')\nWHERE name='students_dict';"
    with engine.begin() as connection:
        connection.execute(text(update_stmt))
    results_CQA,results_with_truth_set_CQA = calculate_expected_values(query_file_name_expected_CQA, query_file_name_actual_CQA, engine)
    results_Prob,results_with_truth_set_prob = calculate_expected_values(query_file_name_expected_Prob, query_file_name_actual_Prob, engine)
    precision_CQA, recall_CQA,coverage_CQA, noise_CQA = summarize_results(results_CQA)
    print(f"\nPrecision CQA: {precision_CQA:.4f}")
    print(f"Recall CQA: {recall_CQA:.4f}")
    print(f"Coverage CQA: {coverage_CQA:.4f}")
    print(f"Noise CQA: {noise_CQA:.4f}")
    precision_prob, recall_prob,coverage_prob, noise_prob = summarize_results(results_Prob)
    print(f"\nPrecision Probabilistic: {precision_prob:.4f}")
    print(f"Recall Probabilistic: {recall_prob:.4f}")
    print(f"Coverage Probabilistic: {coverage_prob:.4f}")
    print(f"Noise Probabilistic: {noise_prob:.4f}")
    total_precision_CQA += precision_CQA
    total_recall_CQA += recall_CQA
    total_coverage_CQA += coverage_CQA
    total_noise_CQA += noise_CQA

    total_precision_prob += precision_prob
    total_recall_prob += recall_prob
    total_coverage_prob += coverage_prob
    total_noise_prob += noise_prob
    
    thresholds, precision_p, recall_p, f1_p = compute_precision_recall_at_thresholds(results_with_truth_set_prob)
    sum_precision_p = [a + b for a, b in zip(sum_precision_p, precision_p)]
    sum_recall_p = [a + b for a, b in zip(sum_recall_p, recall_p)]
    sum_f1_p = [a + b for a, b in zip(sum_f1_p, f1_p)]
```

```python
avg_precision_CQA = total_precision_CQA / len(datasets)
avg_recall_CQA = total_recall_CQA / len(datasets)
avg_coverage_CQA = total_coverage_CQA / len(datasets)
avg_noise_CQA = total_noise_CQA / len(datasets)

avg_precision_prob = total_precision_prob / len(datasets)
avg_recall_prob = total_recall_prob / len(datasets)
avg_coverage_prob = total_coverage_prob / len(datasets)
avg_noise_prob = total_noise_prob / len(datasets)

avg_precision_p = [x / len(datasets) for x in sum_precision_p]
avg_recall_p = [x / len(datasets) for x in sum_recall_p]
avg_f1_p = [x / len(datasets) for x in sum_f1_p]
```
